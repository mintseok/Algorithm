```python
from random import randint, seed

# seed random number generator
seed(1)

array = []
for _ in range(1000):
  array.append(randint(0, 3000))
```

```python
import time

def selectionSort(arr):
  for i in range(len(arr)):
    min = i
    for j in range(len(arr[i:])):
      if arr[min] >= arr[i+j]:
        min = i+j
    if min != i:
      arr[i], arr[min] = arr[min], arr[i]

start = time.time()
selectionSort(array)
end = time.time()

print(f"time taken to SELECTION SORT 1000 data >> {end-start} sec")
```
